```python
# Extract ending year from agricultural year

master["End_Year"]=(
master["Year"]
.str.split("-")
.str[1]
.astype(int)
)

# Convert

master["End_Year"]=master["Start_Year"].floordiv(100)*100+master["End_Year"]

master[["Year","Start_Year","End_Year"]].head()
```

```python
master["Decade"]=(
master["Start_Year"]//10
)*10
```

```python
bins=[1996,2000,2005,2010,2015,2025]

labels=[

"1997-2000",

"2001-2005",

"2006-2010",

"2011-2015",

"2016-2020"

]

master["Year_Group"]=pd.cut(

master["Start_Year"],

bins=bins,

labels=labels

)
```

```python
# ==========================================================
# PHASE 6B
# AGRICULTURE BUSINESS FEATURES
# ==========================================================

import numpy as np

# ----------------------------------------------------------
# Production Per Hectare
# ----------------------------------------------------------
# Measures actual productivity.

master["Production_per_Hectare"] = (
    master["Production"] / master["Area"]
)

# ----------------------------------------------------------
# Fertilizer Efficiency
# ----------------------------------------------------------
# Fertilizer required per hectare.

master["Fertilizer_per_Hectare"] = (
    master["fertilizer"] / master["Area_yield"]
)

# ----------------------------------------------------------
# Pesticide Efficiency
# ----------------------------------------------------------
# Pesticide required per hectare.

master["Pesticide_per_Hectare"] = (
    master["pesticide"] / master["Area_yield"]
)

# ----------------------------------------------------------
# Rainfall Intensity
# ----------------------------------------------------------

master["Rainfall_per_Hectare"] = (
    master["total_rainfall_mm"] / master["Area"]
)
```

```python
master[
[
"Production_per_Hectare",
"Fertilizer_per_Hectare",
"Pesticide_per_Hectare",
"Rainfall_per_Hectare"
]
].head()
```

```python
# ==========================================================
# Crop Category Mapping
# ==========================================================

cereals = [
    "Rice","Wheat","Maize","Barley","Jowar","Bajra","Ragi"
]

pulses = [
    "Gram","Arhar/Tur","Moong","Urad","Masoor"
]

oilseeds = [
    "Groundnut",
    "Soyabean",
    "Mustard",
    "Sunflower",
    "Sesamum"
]

fruits = [
    "Banana",
    "Mango",
    "Apple",
    "Orange",
    "Grapes"
]

vegetables = [
    "Potato",
    "Onion",
    "Tomato",
    "Cabbage"
]

def crop_category(crop):

    crop=str(crop)

    if crop in cereals:
        return "Cereal"

    elif crop in pulses:
        return "Pulse"

    elif crop in oilseeds:
        return "Oilseed"

    elif crop in fruits:
        return "Fruit"

    elif crop in vegetables:
        return "Vegetable"

    else:
        return "Other"

master["Crop_Category"]=master["Crop"].apply(crop_category)
```

```python
master["Crop_Category"].value_counts()
```

```python
# ==========================================================
# Production Level
# ==========================================================

master["Production_Level"] = pd.qcut(

    master["Production"],

    q=3,

    labels=[
        "Low",
        "Medium",
        "High"
    ]

)

# ==========================================================
# Yield Level
# ==========================================================

master["Yield_Level"] = pd.qcut(

    master["Yield"],

    q=3,

    labels=[
        "Low",
        "Medium",
        "High"
    ]

)
```

```python
master[
[
"Production_Level",
"Yield_Level"
]
].head()
```

```python
# ==========================================================
# Weather Available
# ==========================================================

master["Weather_Data_Available"]=(

master["avg_temp_c"].notna()

)

# ==========================================================
# Soil Available
# ==========================================================

master["Soil_Data_Available"]=(

master["N"].notna()

)
```
