```python
# ============================================================
# AgriIntel
#
# Phase 3
#
# Data Cleaning & Preprocessing
# ============================================================

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
crop_df = pd.read_csv("../data/raw/crop_production.csv")
yield_df = pd.read_csv("../data/raw/crop_yield.csv")
weather_df = pd.read_csv("../data/raw/state_weather.csv")
soil_df = pd.read_csv("../data/raw/state_soil.csv")
```

```python
# Create a copy to preserve original data

crop_clean = crop_df.copy()
yield_clean = yield_df.copy()
weather_clean = weather_df.copy()
soil_clean = soil_df.copy()
```

```python
crop_clean.isnull().sum()
```

```python
crop_clean[crop_clean["Crop"].isnull()]
```

```python
crop_clean[crop_clean["Area"].isnull()]
```

```python
crop_clean[crop_clean["Production"].isnull()]
```

```python
# Crop name is mandatory.
# Rows without crop information cannot be analyzed.

crop_clean.dropna(subset=["Crop"], inplace=True)
```

```python
crop_clean.dropna(subset=["Season"], inplace=True)
```

```python
crop_clean["Area"].isnull().sum()
```

```python
crop_clean["Area"] = crop_clean["Area"].fillna(
    crop_clean["Area"].median()
)
```

```python
crop_clean["Yield"] = crop_clean["Yield"].fillna(
    crop_clean["Yield"].median()
)
```

```python
crop_clean["Production"].isnull().sum()

# Production is the target variable of the project. Missing production values should not be imputed without domain knowledge because it may introduce bias.
```

```python
crop_clean = crop_clean.drop_duplicates()
```

```python
text_columns = [
    "State",
    "District",
    "Crop",
    "Season"
]

for col in text_columns:
    crop_clean[col] = (
        crop_clean[col]
        .str.strip()
        .str.title()
    )
```

```python
yield_clean.columns
```

```python
yield_clean["crop"] = yield_clean["crop"].str.strip().str.title()

yield_clean["state"] = yield_clean["state"].str.strip().str.title()

yield_clean["season"] = yield_clean["season"].str.strip().str.title()
```

```python
weather_clean["state"] = weather_clean["state"].str.strip().str.title()
```

```python
soil_clean["state"] = soil_clean["state"].str.strip().str.title()
```

```python
crop_clean.to_csv(
    "../data/processed/crop_production_clean.csv",
    index=False
)

yield_clean.to_csv(
    "../data/processed/crop_yield_clean.csv",
    index=False
)

weather_clean.to_csv(
    "../data/processed/state_weather_clean.csv",
    index=False
)

soil_clean.to_csv(
    "../data/processed/state_soil_clean.csv",
    index=False
)
```
